# Soal 3: DataFrame vs Series

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

1. A. SERIES 
- Merupakan satu dimensi yang dapat menyimpan berbagai jenis tipe data dan  seragam
- Gabungan antara list dan dictionary
- Dapat dilakukan metode slicing yang mirip dengan list
- Index yang dimiliki pada series berguna sebagai keys pada dictionary

B. LIST
- Struktur data yang paling dinamis daripada yang lain.
- Termasuk jenis struktur data yang anggotanya dapat diubah (ditambah, dikurang/dihapus, dibagi) sesuai dengan keinginan dan kebutuhan.
- List dapat diakses menggunakan index
- List menggunakan metode last in first out.
- Dibatasi oleh tanda siku []

C. DICTIONARY
- Struktur data yang anggotanya berpasangan (key:value)
- Struktur data yang dibatasi dengan tanda kurung kurawal {}
- Struktur data yang anggotanya bisa diubah
- Struktur data yang tidak dapat diakses dengan index
- Struktur data yang dapat diakses nilai dengan kunci

2. A. DATAFRAME
- Merupakan suatu objek 2 dimensi tempat menyimpan data yang lebih terstruktur
- Memiliki 2 index, yaitu index baris dan index columns
- Pada satu columns dataframe harus memiliki tipe data yang sama
- Namun, antar columns dataframe bisa memiliki jenis data yang berbeda

B. SERIES
- Merupakan satu dimensi yang dapat menyimpan berbagai jenis tipe data dan  seragam
- Gabungan antara list dan dictionary
- Dapat dilakukan metode slicing yang mirip dengan list
- Index yang dimiliki pada series berguna sebagai keys pada dictionary

```python
#Buatlah suatu data series

import pandas as pd

x = pd.Series([7,5,3,4,7])
print(x)

x.index = ['a', 'b', 'c', 'd', 'e']
print(x)
```

```python
# Persiapan membuat data
import pandas as pd
import random

pelajaran = ['Matematika', 'Fisika', 'Biologi', 'Kimia', 'Geologi', 'Ekonomi']
kelas = ['12A', '12B', '12C', '12D']

hasil_nilai = {
    'siswa_id' : range(1, 21),
    'pelajaran' : [random.choice(pelajaran) for i in range(20)],
    'nilai' : [random.choice(range(0, 100)) for i in range(20)],
    'jenis-kelamin' : [random.choice(['laki-laki', 'perempuan']) for i in range(20)],
    'kelas' : [random.choice(kelas) for i in range(20)]
}

df = pd.DataFrame(hasil_nilai)
df.head()
```

# Soal 6: Filter DataFrame Dengan Boolean

1. Tampilkan semua data siswa pada mata pelajaran Ekonomi dan Fisika
2. Tampilkan semua data siswa yang mempunyai nilai lebih dari 50 di mata pelajaran Kimia
3. Tampilkan semua data yang mempunyai nilai lebih dari 70, berjnis kelamin perempuan dan di kelas 12C

```python
display(df[(df.pelajaran=='Ekonomi') | (df.pelajaran=='Fisika')])
```

```python
display(df[(df.pelajaran=='Kimia') & (df.nilai>50)])
```

```python
display(df[(df.nilai<70) & (df['jenis-kelamin']=='perempuan') & (df.kelas=='12C')])
```
